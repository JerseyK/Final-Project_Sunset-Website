```python
!pip

import numpy as np
import pandas as pd
# import matplotlib as plt
# import seaborn as sns
import plotly.express as px
# import plotly.graph_objects as go
# import plotly.io as pio
#import streamlit as st
import altair as alt
# from st_aggrid import AgGrid, GridOptionsBuilder
# from st_aggrid.shared import GridUpdateMode

#from update_data_cache import get_data

#from pypfopt.efficient_frontier import EfficientFrontier

#pio.renderers.default='browser' # use when doing dev in Spyder (to show figs)

# Page config
st.set_page_config(
    "ADD INFO HERE! by Sunset",
    "ðŸ“Š",
    layout="wide",
)


###############################################################################
# starting: "main" page of dashboard
###############################################################################
st.title('MATCH TITLE TO TAB TITLE OR MAKE IT *Overall Seller Distrubution*') 
# below is another way to do the headers/write text can either use """ or ''' before & after the text
#  """
# # Header 1 (can make this a)
# ## subhearder
# """

#############################################
# Plots side by side
#############################################
#@st.cache_data
##########################################,,,
##ANOTHER WAY FOR MAKING THE GRAPHS
# Load data
df = pd.read_csv('data/merge.csv')

# Convert srcdate to a datetime column
df['srcdate'] = pd.to_datetime(df['srcdate'])

# Create a year column
df['year'] = df['srcdate'].dt.year

# Create a bar chart
@st.cache
def create_bar_chart(df):
    chart = alt.Chart(df).mark_bar().encode(
        x='GICS Sector',
        y='salecs',
        color='ctype',
        column='year'
    ).properties(
        width=300,
        height=300,
        title='Sales by GICS Sector and Ctype'
    )
    return chart

# Group the data by ctype, GICS Sector, and year
grouped_df = df.groupby(['ctype', 'GICS Sector', 'year'])['salecs'].sum().reset_index()

# Normalize sales to get percentages
grouped_df['sales'] = grouped_df.groupby(['ctype', 'year'])['salecs'].apply(lambda x: x/x.sum())

# Create the bar chart
chart = create_bar_chart(grouped_df)

# Display the chart
st.altair_chart(chart)
```
